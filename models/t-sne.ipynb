```python
import pandas as pd
import numpy as np
from sklearn.manifold import TSNE
from matplotlib import pyplot as plt
plt.rcParams['figure.figsize'] = [12, 8]
plt.rcParams['figure.dpi'] = 100
```

```python
df = pd.read_csv('../data/men_lead_lag7.csv', index_col=0)
df[['height', 'weight', 'age']] = df[['height', 'weight', 'age']].fillna(df[df['Year'] == 2019].mean())
df[['t-1', 't-2', 't-3', 't-4', 't-5', 't-6', 't-7']] = df[['t-1', 't-2', 't-3', 't-4', 't-5', 't-6', 't-7']].fillna(0)

X1 = df.loc[df['Year'] == 2019,['height', 'weight', 'age', 't-1', 't-2', 't-3', 't-4', 't-5', 't-6', 't-7']].to_numpy()
for i, p in enumerate([2, 5, 10, 25, 50], 1):
    X_answer1 = tsne = TSNE(perplexity=p).fit_transform(X1)
    plt.subplot(2, 3, i)
    plt.scatter(X_answer1[:,0], X_answer1[:,1])
    plt.title(f'$p = {p}$')
plt.suptitle('With height, weight, and age')
plt.tight_layout()
plt.show()
```

```python
X2 = df.loc[df['Year'] == 2019,['t-1', 't-2', 't-3', 't-4', 't-5', 't-6', 't-7']].to_numpy()
for i, p in enumerate([2, 5, 10, 25, 50], 1):
    X_answer2 = tsne = TSNE(perplexity=p).fit_transform(X2)
    plt.subplot(2, 3, i)
    plt.scatter(X_answer2[:,0], X_answer2[:,1])
    plt.title(f'$p = {p}$')
plt.suptitle('Without height, weight, and age')
plt.tight_layout()
plt.show()
```
